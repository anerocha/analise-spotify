# **Projeto: Análise de Preferências Musicais e Recomendação de Artistas**

**Objetivo:**
Analisar uma base de dados de músicas populares do Spotify para entender quais são os padrões de preferência entre essas músicas, incluindo artistas, gêneros e características de áudio, como popularidade, energia e dançabilidade. Com esses dados, o projeto fornecerá uma visão abrangente sobre o que torna uma música popular e poderá sugerir músicas semelhantes com base em características favoritas.

### **Etapas do Projeto:**

**Carregamento dos Dados:** Importar o arquivo CSV universal_top_spotify_songs.csv, que contém informações sobre músicas populares no Spotify, incluindo nome, artista, gênero, popularidade e características de áudio.

**Limpeza e Preparação:** Remover duplicatas e tratar valores ausentes para garantir a precisão dos dados.

**Análise Exploratória de Dados (EDA):** Examinar a distribuição dos dados e explorar preferências musicais, focando em:

*   Artistas e gêneros mais populares.
*   Características de áudio (popularidade, energia, dançabilidade).

**Análise de Preferências Musicais:** Comparar características das músicas para entender as relações entre popularidade, energia e dançabilidade.

**Recomendações de Músicas:** Sugerir músicas com base nas características mais valorizadas pelos ouvintes.


### **Resultados Esperados:**


*  Identificação de gêneros e artistas populares.
*   Visualização das distribuições de popularidade, energia e dançabilidade.
*   Identificação de características comuns entre músicas populares.
*   Sugestões de músicas com base nas características desejadas, visando criar playlists personalizadas.



## 1\.Carregar os Dados

```python
import pandas as pd
from google.colab import files

# Fazer o upload do arquivo CSV
uploaded = files.upload()

file_path = 'universal_top_spotify_songs.csv'

# Carregar a base de dados CSV
spotify_data = pd.read_csv(file_path)

# Exibir as primeiras linhas do DataFrame
print("Visualização inicial dos dados:")
print(spotify_data.head())
```

## 2\.Limpeza e Preparação dos Dados

```python
# Remover duplicatas e valores ausentes
spotify_data.drop_duplicates(inplace=True)
spotify_data.dropna(inplace=True)

print(f"Após limpeza: {spotify_data.shape[0]} registros restantes.")
```

## 3.Análise Exploratória dos Dados (EDA)

Vamos visualizar os artistas e gêneros mais populares, bem como explorar as características de áudio (como popularidade e energia).

### **3.1 Artistas e Gêneros Mais Populares**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Top 10 artistas mais frequentes
top_artists = spotify_data['artists'].value_counts().head(10)
plt.figure(figsize=(9, 6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette="viridis")
plt.title("Top 10 Artistas Mais Frequentes")
plt.xlabel("Número de Ocorrências")
plt.ylabel("Artista")
plt.show()

```

```python
# Top 10 álbuns mais populares
top_albums = spotify_data['album_name'].value_counts().head(10)
plt.figure(figsize=(9, 6))
sns.barplot(x=top_albums.values, y=top_albums.index, palette="plasma")
plt.title("Top 10 Álbuns Mais Populares")
plt.xlabel("Número de Ocorrências")
plt.ylabel("Álbum")
plt.show()
```

```python
# Distribuição de popularidade
plt.figure(figsize=(9, 6))
sns.histplot(spotify_data['popularity'], kde=True, color="blue")
plt.title("Distribuição da Popularidade das Músicas")
plt.xlabel("Popularidade")
plt.ylabel("Frequência")
plt.show()
```

## 4.Análise de Preferências Musicais com Base nas Características

Aqui, vamos examinar como características como dançabilidade e energia variam entre as músicas mais populares.

```python
# Gráfico de dispersão para Popularidade vs Dançabilidade
plt.figure(figsize=(9, 6))
sns.scatterplot(data=spotify_data, x='danceability', y='popularity', alpha=0.7)
plt.title("Popularidade vs Dançabilidade")
plt.xlabel("Dançabilidade")
plt.ylabel("Popularidade")
plt.show()

# Gráfico de dispersão para Energia vs Popularidade
plt.figure(figsize=(9, 6))
sns.scatterplot(data=spotify_data, x='energy', y='popularity', alpha=0.7)
plt.title("Popularidade vs Energia")
plt.xlabel("Energia")
plt.ylabel("Popularidade")
plt.show()
```

## 5.Recomendações Baseadas nas Preferências

Para simular recomendações, selecionaremos músicas com características de popularidade, energia e dançabilidade semelhantes às das músicas mais populares do conjunto de dados.

```python
# Definindo limites de popularidade, energia e dançabilidade para recomendações
popularity_threshold = spotify_data['popularity'].quantile(0.75)
energy_threshold = spotify_data['energy'].quantile(0.75)
danceability_threshold = spotify_data['danceability'].quantile(0.75)

# Filtrando músicas recomendadas com base nas preferências
recommended_songs = spotify_data[
    (spotify_data['popularity'] >= popularity_threshold) &
    (spotify_data['energy'] >= energy_threshold) &
    (spotify_data['danceability'] >= danceability_threshold)
]

# Exibir as músicas recomendadas, removendo duplicatas com base em 'name' e 'artists'
try:
    print("Músicas Recomendadas Baseadas nas Preferências:")
    recommended_unique = recommended_songs[['name', 'artists', 'popularity', 'energy', 'danceability']].drop_duplicates(subset=['name', 'artists'])
    print(recommended_unique.head(10))
except KeyError as e:
    print(f"Erro: {e}. Verifique os nomes das colunas.")
```
